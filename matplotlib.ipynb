## **ðŸ“Œ 1. GrÃ¡fico de Linhas (plot)**

ðŸ“ˆ Usado para representar dados contÃ­nuos, como tendÃªncias ao longo do tempo.

```python
#grÃ¡fico de linha

import matplotlib.pyplot as plt
import numpy as np


x2 = np.linspace(0, 10, 1000)
y2 = np.cos(x2)

x3 = np.linspace(0, 10, 1000)
y3 = np.sin(x3)

plt.plot(x2, y2, label="cosseno")
plt.plot(x3, y3, label="seno")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("GrÃ¡fico de Linhas")
plt.legend()
plt.show()
```

## **ðŸ“Œ 2. GrÃ¡fico de DispersÃ£o (scatter)**

ðŸ”µ Usado para visualizar a relaÃ§Ã£o entre duas variÃ¡veis.

```python
x1 = np.random.rand(10)
y1 = np.random.rand(10)

x2 = np.random.rand(10)
y2 = np.random.rand(10)

x3 = np.random.rand(10)
y3 = np.random.rand(10)

x4 = np.random.rand(10)
y4 = np.random.rand(10)
plt.scatter(x1, y1, color='red', marker='o')
plt.scatter(x2, y2, color='blue', marker='x')
plt.scatter(x3, y3, color='green', marker='^')
plt.scatter(x4, y4, color='yellow', marker='s')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("GrÃ¡fico de DispersÃ£o")
plt.show()
```

## **ðŸ“Œ 3. GrÃ¡fico de Barras (bar)**

ðŸ“Š Ãštil para comparar categorias.

```python
categorias = ["A", "B", "C", "D"]
valores = [4, 7, 1, 8]

plt.bar(categorias, valores, color="lightpink")
plt.xlabel("Categorias")
plt.ylabel("Valores")
plt.title("GrÃ¡fico de Barras")
plt.show()
```

## **ðŸ“Œ 4. Histograma (hist)**

ðŸ“‰ Usado para visualizar distribuiÃ§Ãµes de dados.

```python
dados = np.random.randn(1000)

plt.hist(dados, bins=30, color="lightgreen", edgecolor="black")
plt.xlabel("Valores")
plt.ylabel("FrequÃªncia")
plt.title("Histograma")
plt.show()
```

## **ðŸ“Œ 5. GrÃ¡fico de Pizza (pie)**

ðŸ¥§ Para representar proporÃ§Ãµes.

```python
labels = ["A", "B", "C", "D"]
valores = [20, 30, 25, 25]

plt.pie(valores, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("GrÃ¡fico de Pizza")
plt.show()
```

## **ðŸ“Œ 6. GrÃ¡fico de Ãrea (fill_between)**

ðŸ“‰ Similar ao de linha, mas preenche a Ã¡rea abaixo da curva.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.fill_between(x, y1, y2, color="lightblue", alpha=0.5)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("GrÃ¡fico de Ãrea")
plt.show()
```

## **ðŸ“Œ 7. Boxplot (Diagrama de Caixa) (boxplot)**

ðŸ“¦ Ãštil para visualizar distribuiÃ§Ãµes estatÃ­sticas e valores discrepantes.

```python
dados = [np.random.rand(50) * 10 for _ in range(4)]

plt.boxplot(dados, labels=["A", "B", "C", "D"])
plt.xlabel("Categorias")
plt.ylabel("Valores")
plt.title("Boxplot")
plt.show()
```

## **ðŸ“Œ 8. GrÃ¡fico de Contorno (contour)**

ðŸŒ„ Representa funÃ§Ãµes 3D em 2D usando curvas de nÃ­vel.

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)

plt.contour(X, Y, Z, levels=20, cmap="viridis")
plt.colorbar()
plt.title("GrÃ¡fico de Contorno")
plt.show()
```

## **ðŸ“Œ 9. GrÃ¡fico de SuperfÃ­cie (plot_surface) (NecessÃ¡rio mpl_toolkits.mplot3d)**

ðŸŒ Para visualizaÃ§Ãµes 3D.

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)

ax.plot_surface(X, Y, Z, cmap="viridis")
plt.title("GrÃ¡fico de SuperfÃ­cie 3D")
plt.show()
```

## **ðŸ“Œ 10. GrÃ¡fico de Erros (errorbar)**

ðŸ“‰ Ãštil para dados com incerteza.

```python
x = np.linspace(0, 10, 10)
y = np.sin(x)
erro = 0.2

plt.errorbar(x, y, yerr=erro, fmt="o", color="red", capsize=5)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("GrÃ¡fico de Erros")
plt.show()
```
