## **📌 1. Gráfico de Linhas (plot)**

📈 Usado para representar dados contínuos, como tendências ao longo do tempo.

```python
#gráfico de linha

import matplotlib.pyplot as plt
import numpy as np


x2 = np.linspace(0, 10, 1000)
y2 = np.cos(x2)

x3 = np.linspace(0, 10, 1000)
y3 = np.sin(x3)

plt.plot(x2, y2, label="cosseno")
plt.plot(x3, y3, label="seno")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Gráfico de Linhas")
plt.legend()
plt.show()
```

## **📌 2. Gráfico de Dispersão (scatter)**

🔵 Usado para visualizar a relação entre duas variáveis.

```python
x1 = np.random.rand(10)
y1 = np.random.rand(10)

x2 = np.random.rand(10)
y2 = np.random.rand(10)

x3 = np.random.rand(10)
y3 = np.random.rand(10)

x4 = np.random.rand(10)
y4 = np.random.rand(10)
plt.scatter(x1, y1, color='red', marker='o')
plt.scatter(x2, y2, color='blue', marker='x')
plt.scatter(x3, y3, color='green', marker='^')
plt.scatter(x4, y4, color='yellow', marker='s')
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Gráfico de Dispersão")
plt.show()
```

## **📌 3. Gráfico de Barras (bar)**

📊 Útil para comparar categorias.

```python
categorias = ["A", "B", "C", "D"]
valores = [4, 7, 1, 8]

plt.bar(categorias, valores, color="lightpink")
plt.xlabel("Categorias")
plt.ylabel("Valores")
plt.title("Gráfico de Barras")
plt.show()
```

## **📌 4. Histograma (hist)**

📉 Usado para visualizar distribuições de dados.

```python
dados = np.random.randn(1000)

plt.hist(dados, bins=30, color="lightgreen", edgecolor="black")
plt.xlabel("Valores")
plt.ylabel("Frequência")
plt.title("Histograma")
plt.show()
```

## **📌 5. Gráfico de Pizza (pie)**

🥧 Para representar proporções.

```python
labels = ["A", "B", "C", "D"]
valores = [20, 30, 25, 25]

plt.pie(valores, labels=labels, autopct="%1.1f%%", startangle=90)
plt.title("Gráfico de Pizza")
plt.show()
```

## **📌 6. Gráfico de Área (fill_between)**

📉 Similar ao de linha, mas preenche a área abaixo da curva.

```python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.fill_between(x, y1, y2, color="lightblue", alpha=0.5)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Gráfico de Área")
plt.show()
```

## **📌 7. Boxplot (Diagrama de Caixa) (boxplot)**

📦 Útil para visualizar distribuições estatísticas e valores discrepantes.

```python
dados = [np.random.rand(50) * 10 for _ in range(4)]

plt.boxplot(dados, labels=["A", "B", "C", "D"])
plt.xlabel("Categorias")
plt.ylabel("Valores")
plt.title("Boxplot")
plt.show()
```

## **📌 8. Gráfico de Contorno (contour)**

🌄 Representa funções 3D em 2D usando curvas de nível.

```python
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)

plt.contour(X, Y, Z, levels=20, cmap="viridis")
plt.colorbar()
plt.title("Gráfico de Contorno")
plt.show()
```

## **📌 9. Gráfico de Superfície (plot_surface) (Necessário mpl_toolkits.mplot3d)**

🌍 Para visualizações 3D.

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

X, Y = np.meshgrid(x, y)
Z = np.sin(X**2 + Y**2)

ax.plot_surface(X, Y, Z, cmap="viridis")
plt.title("Gráfico de Superfície 3D")
plt.show()
```

## **📌 10. Gráfico de Erros (errorbar)**

📉 Útil para dados com incerteza.

```python
x = np.linspace(0, 10, 10)
y = np.sin(x)
erro = 0.2

plt.errorbar(x, y, yerr=erro, fmt="o", color="red", capsize=5)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Gráfico de Erros")
plt.show()
```
